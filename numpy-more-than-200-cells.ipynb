You can see below while i changed the new_arr index 0 to 100, the old one also got changed. 
this is because of array broadcasting

```python
arr = np.arange(10)
print(arr)

new_arr = arr[:4]
print(new_arr)

new_arr[0] = 100
print(new_arr)
print(arr)
```

**To Avoid that use copy()**

```python
arr = np.arange(10)
print(arr)

new_arr = arr[:4].copy()
print(new_arr)

new_arr[0] = 100
print(new_arr)
print(arr)
```

## Array Conditionals

```python
arr = np.arange(10)
print(arr)

print(arr>4)
print(arr[arr>4])
```

```python
#Exercise: get even numbers from arr = np.arange(10)
print(arr[arr%2 ==0])

# Values divisible by 2 and 3

print(arr[(arr%2 ==0) & (arr%3 ==0)])

# Values divisible by 2 or 3

print(arr[(arr%2 ==0) | (arr%3 ==0)])
```

## Stacking

```python
# Which one is vertical
# 1 
# 2-- 

arr1 = np.random.randint(0,999,(2,3))
arr2 = np.random.randint(0,999,(2,3))

print(arr1,arr2)
```

```python
# Vertical stack >> No of columns should be same

print(np.vstack([arr1,arr2]))
```

```python
# Horizontal Stack >> No of rows should be same

print(np.hstack([arr1,arr2]))
```
